```python
# Identify columns with non-numeric values
non_numeric_columns = application_data.select_dtypes(include=['object']).columns
print("Columns with non-numeric values:\n", non_numeric_columns)
```

```python
# Convert categorical columns to numeric using one-hot encoding
application_data_encoded = pd.get_dummies(application_data, columns=non_numeric_columns)
non_numeric_columns_previous_app = previous_application.select_dtypes(include=['object']).columns
previous_application_encoded = pd.get_dummies(previous_application, columns=non_numeric_columns_previous_app)

# Display the first few rows of the encoded dataframes
# print(application_data_encoded.head())
# print(previous_application_encoded.head())
```

```python
# Identify missing data in encoded application_data
missing_data_application_encoded = application_data_encoded.isnull().sum()
missing_data_application_encoded = missing_data_application_encoded[missing_data_application_encoded > 0]
missing_data_application_encoded_percent = (missing_data_application_encoded / len(application_data_encoded)) * 100
print("Missing data in encoded application_data:\n", missing_data_application_encoded_percent)

# Fill missing values in encoded dataframes (example: median for numerical columns)
application_data_encoded = application_data_encoded.fillna(application_data_encoded.median())
previous_application_encoded = previous_application_encoded.fillna(previous_application_encoded.median())
```

```python
application_data = application_data_encoded
previous_application = previous_application_encoded
```

## PLOTLY CODE

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Handle missing values
application_data.fillna(application_data.mean(), inplace=True)
previous_application.fillna(previous_application.mean(), inplace=True)

# Basic information about the datasets
print(application_data.info())
print(previous_application.info())

# Distribution of the target variable
fig1 = px.histogram(application_data, x='TARGET', title='Distribution of Target Variable')
fig1.show()

# Distribution of income
fig2 = px.histogram(application_data, x='AMT_INCOME_TOTAL', nbins=50, title='Distribution of Income')
fig2.show()

# Distribution of loan amounts
fig3 = px.histogram(application_data, x='AMT_CREDIT', nbins=50, title='Distribution of Loan Amounts')
fig3.show()

# Loan type vs. Loan amount
fig4 = px.box(application_data, x='NAME_CONTRACT_TYPE', y='AMT_CREDIT', title='Loan Type vs Loan Amount')
fig4.show()

# Gender vs. Target
fig5 = px.histogram(application_data, x='CODE_GENDER', color='TARGET', barmode='group', title='Gender vs Target')
fig5.show()

# Income vs. Target
fig6 = px.box(application_data, x='TARGET', y='AMT_INCOME_TOTAL', title='Income vs Target')
fig6.show()

# Education type vs. Target
fig7 = px.histogram(application_data, x='NAME_EDUCATION_TYPE', color='TARGET', barmode='group', title='Education Type vs Target')
fig7.show()

# Occupation type vs. Target
fig8 = px.histogram(application_data, x='OCCUPATION_TYPE', color='TARGET', barmode='group', title='Occupation Type vs Target')
fig8.show()

# External sources vs. Target
fig9 = make_subplots(rows=1, cols=3, subplot_titles=('EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3'))

fig9.add_trace(go.Box(y=application_data['EXT_SOURCE_1'], x=application_data['TARGET'], name='EXT_SOURCE_1'), row=1, col=1)
fig9.add_trace(go.Box(y=application_data['EXT_SOURCE_2'], x=application_data['TARGET'], name='EXT_SOURCE_2'), row=1, col=2)
fig9.add_trace(go.Box(y=application_data['EXT_SOURCE_3'], x=application_data['TARGET'], name='EXT_SOURCE_3'), row=1, col=3)

fig9.update_layout(title_text='External Sources vs Target')
fig9.show()

# Previous application status
fig10 = px.histogram(previous_application, x='NAME_CONTRACT_STATUS', title='Previous Application Status')
fig10.show()

# Previous loan amount
fig11 = px.histogram(previous_application, x='AMT_CREDIT', nbins=50, title='Previous Loan Amount')
fig11.show()

# Previous application purpose
fig12 = px.histogram(previous_application, x='NAME_CASH_LOAN_PURPOSE', title='Previous Application Purpose')
fig12.show()

# Distribution of previous application days
fig13 = px.histogram(previous_application, x='DAYS_DECISION', nbins=50, title='Days of Previous Application Decision')
fig13.show()

# Loan type vs. Amount requested
fig14 = px.box(previous_application, x='NAME_CONTRACT_TYPE', y='AMT_APPLICATION', title='Loan Type vs Amount Requested')
fig14.show()
```
